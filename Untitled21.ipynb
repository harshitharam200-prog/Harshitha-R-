```python
import random

def fishing_game():
    """
    A simple fishing game where players catch random fish with specified values.
    """

    # Define fish types and their values using a dictionary
    fish_types = {
        "Tuna": 10,
        "Salmon": 15,
        "Cod": 8,
        "Shark": 50,
        "Sardine": 2,
        "Eel": 5,
        "Pufferfish": 25,
        "Boot": 0, # Sometimes you catch trash!
        "Old Tire": 0
    }

    total_score = 0
    inventory = {} # Initialize an empty dictionary for the inventory
    print("Welcome to the Fishing Game!")
    print("Catch fish to earn points. Each fish has a different value.")
    print("Type 'fish' to cast your line, 'inventory' to view your catch, or 'quit' to end the game.\n")

    while True:
        action = input("What do you want to do? (fish/inventory/quit): ").lower()

        if action == 'fish':
            # Randomly select a fish
            caught_fish_name = random.choice(list(fish_types.keys()))
            caught_fish_value = fish_types[caught_fish_name]

            total_score += caught_fish_value

            # Update inventory
            inventory[caught_fish_name] = inventory.get(caught_fish_name, 0) + 1

            if caught_fish_value > 0:
                print(f"You cast your line and caught a {caught_fish_name}! (+{caught_fish_value} points)")
            else:
                print(f"You cast your line and caught an {caught_fish_name}... no points for that one!")

            print(f"Your current score: {total_score}\n")
        elif action == 'inventory':
            print("\n--- Your Inventory ---")
            if not inventory:
                print("Your inventory is empty. Go catch some fish!")
            else:
                for item, count in inventory.items():
                    print(f"- {item}: {count}")
            print("----------------------\n")
        elif action == 'quit':
            print(f"Thanks for playing! Your final score is: {total_score}")
            print("Final Inventory:")
            if not inventory:
                print("No fish caught!")
            else:
                for item, count in inventory.items():
                    print(f"- {item}: {count}")
            break
        else:
            print("Invalid action. Please type 'fish', 'inventory', or 'quit'.\n")

# Run the game
fishing_game()
```
