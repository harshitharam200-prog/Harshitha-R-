```python
import matplotlib.pyplot as plt
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
production_start=[12324,43567,78964,56748,45675,23445,23557,64743,24347,45366]
production_close=[1233,345,345,5667,4557,4567,456,3456,5467,46766]
ind=np.arange(len(years))
width=0.3
plt.bar(ind,production_start,width,label="production_start")
plt.bar(ind+width,production_close,width,label="production_close",color="g")
plt.title("production per year graph")
plt.xlabel("years")
plt.ylabel("production")
plt.xticks (ind,years,rotation=70)
plt.legend()
plt.show()
```
