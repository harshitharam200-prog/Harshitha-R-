```python
keys=['movie','hero','release year']
values=["kantara","rishab",2022]
create_dict=dict(zip(keys,values))
print(create_dict)
```

```python
keys=['movie','hero','release year']
values=["kantara","rishab",2022]
for i in range(len(keys)):
    create_dict[keys[i]]=values[i]
print(create_dict)
```

```python
dict1={"name":"john",10:[2,4,3]}
dict2=dict1
print(dict1)
print(dict2)
print()
print("after change")
print()
dict1.popitem()
print(dict1)
print(dict2)
```

```python
dict1={"name":"john",10:[2,4,3]}
dict2=dict1.copy()
print(dict1)
print(dict2)
print()
print("after change")
print()
dict1.popitem()
print(dict1)
print(dict2)
print("change in dict1 is not reflected in dict2")
```
