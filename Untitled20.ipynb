```python
import random
inventory={}

amount=0
fish_types = {"Tuna": 10, "Salmon": 15,"rohu": 8,"mrigal": 50, "Sardine": 2}
print("lets start the game")
print("1.go fishing")
print("2.view inventory")
print("3.sell fish")
print("4.end game")
while True:
  choice=int(input("enter your choice"))
  if choice==1:
    caught_fish=random.choice(list(fish_types.keys()))
    inventory[caught_fish]=inventory.get(caught_fish,0)+1
    print(f"You caught a {caught_fish}!")

  elif choice==2:
    if inventory=={}:
      print("No fish caught yet.")
    else:
      print("Your inventory:")
      for fish,quantity in inventory.items():
        print(f"{fish}: {quantity}")

  elif choice==3:
    if inventory=={}:
      print("No fish to sell!")
    else:
      earnings=0
      print("Selling fish")
      for fish,quantity in list(inventory.items()):

        earnings+=fish_types[fish]*quantity
        print(f"Sold {quantity} {fish}(s) for ${fish_types[fish]*quantity}.")
      amount+=earnings
      inventory.clear()
      print(f"You sold all your fish for ${earnings}. Total money: ${amount}")

  elif choice==4:
    print(f"Final money earned: ${amount}")
    print("Thanks for playing!")
    break
  else:
    print("Invalid choice. Try again.")
```
