```python
import pandas as pd
data={
    "farm":["farm a","farm b","farm c","farm d"],
    "crop":["wheat","rice","wheat","maize"],
    "area_hectares":[10,15,8,12],
    "prouction_tonne":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.describe() #basic summary in statistic form
```

```python
#add a new column (yield)
#yield=production÷area
df["yield_tonne_per_hectare"]=df["prouction_tonne"]/df["area_hectares"]
print(df)
```

```python
#filter data
#farms growing wheat
print(df["crop"]=="wheat")
print()
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
#farms with yield greater than 3
high_yield=df[df["yield_tonne_per_hectare"]>3]
print(high_yield)
```

```python
#sort the farms by highest production
sorted_df=df.sort_values(by="prouction_tonne",ascending=False)
print(sorted_df)
```

```python
#groupBy operation -average yield per crop
avg_yield=df.groupby("crop")["yield_tonne_per_hectare"].mean()
print(avg_yield)
```
