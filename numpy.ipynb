```python
import numpy as np
crop_yields=np.array([
    [12.2,23.2,34.3,33.2,24.2],
    [10.2,14.3,24.4,23.4,33.3],
    [24.5,66.1,21.1,23.2,22.2],
    [33.1,25.3,12.43,32.1,22.1],
])
print("crop yields(tons)",crop_yields)
print(f"\nshape:{crop_yields.shape}")
print(f"data type:{crop_yields.dtype}")
print(f"total elements:{crop_yields.size}")
print(f"number of dimensions:{crop_yields.ndim}")
```

```python
import numpy as np
total_yield=np.sum(crop_yields)
avg_yield_per_farm=np.mean(crop_yields,axis=1) #row average
avg_yield_per_season=np.mean(crop_yields,axis=0) #column average
max_yield=np.max(crop_yields)
min_yield=np.min(crop_yields)
print(f"total yield:{total_yield}")
print(f"average yield per farm:{avg_yield_per_farm}")
print(f"average yield per season:{avg_yield_per_season}")
print(f"maximum yield:{max_yield}")
print(f"minimum yield:{min_yield}")
```

```python
import numpy as np
best_farm=np.argmax(avg_yield_per_farm) #argmax doesnt return the value , it returns wherre it is located(index)
worst_farm=np.argmin(avg_yield_per_farm)
print(f"best farm:{best_farm}")
print(f"worst farm:{worst_farm}")
```
